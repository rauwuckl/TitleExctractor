```python
from sklearn.cluster import MeanShift, AgglomerativeClustering
import numpy
import matplotlib.pyplot as plt
from utils import *

%matplotlib inline
```

```python
tree = load_tree("data/110000235_datacorr.xml")

pages = tree.getroot()
p4 = list(pages.iter("page"))[1]

tL, _, _, _ = get_coordinates_of_textboxes(p4.findall("./textbox"))
```

```python
clustering = AgglomerativeClustering(distance_threshold=15, n_clusters=None)
```

```python
c = clustering.fit_predict(np.expand_dims(tL, axis=-1))
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
plt.scatter(tL, np.ones_like(tL)+np.random.rand(len(tL)), marker="x", c=c)
```
